```python
import pandas as pd
```

```python
file_path = 'ECOMM DATA.xlsx'
sales_data = pd.read_excel(file_path)
```

```python
sales_data.head()
```

```python
sales_data.info()
```

```python
sales_data.describe()
```

```python
sales_data.isnull().sum()
```

```python
sales_data_cleaned = sales_data.dropna()
```

```python
sales_data_cleaned.info()
```

```python
sales_data_cleaned['Total Sales'] = sales_data_cleaned['Quantity'] * sales_data_cleaned['Profit']
```

```python
sales_data_cleaned.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
sales_trend = sales_data_cleaned.groupby('Order Date')['Total Sales'].sum()
```

```python
sales_trend_rolling = sales_trend.rolling(window=7).mean()
```

```python
plt.figure(figsize=(14, 8))
plt.plot(sales_trend.index, sales_trend.values, label='Daily Sales', color='blue', alpha=0.6)
plt.plot(sales_trend_rolling.index, sales_trend_rolling.values, label='7-Day Rolling Average', color='red', linewidth=2)
plt.axvline(pd.Timestamp('2023-11-24'), color='green', linestyle='--', linewidth=2, label='Black Friday')
plt.text(pd.Timestamp('2023-11-24'), max(sales_trend), 'Black Friday', color='green', fontsize=12, ha='right')
max_sales_date = sales_trend.idxmax()
max_sales_value = sales_trend.max()

plt.annotate(f'Highest Sales: ${max_sales_value:.2f}', xy=(max_sales_date, max_sales_value),
             xytext=(max_sales_date, max_sales_value + max_sales_value * 0.1),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=12, ha='center', color='black')

plt.title('Sales Trend Over Time', fontsize=18)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Total Sales ($)', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
best_selling_products = sales_data_cleaned.groupby('Product Name')['Total Sales'].sum().sort_values(ascending=False)
best_selling_products.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
top_products = best_selling_products.head(10)
```

```python
plt.figure(figsize=(10, 6))
bars = plt.barh(top_products.index, top_products.values, color='pink')

for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, 
             f'${bar.get_width():,.2f}', 
             va='center', ha='left', color='black', fontsize=10)

plt.title('Top 10 Best-Selling Products', fontsize=16, fontweight='bold')
plt.xlabel('Total Sales', fontsize=14)
plt.ylabel('Product', fontsize=14)
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

```python
sales_by_category = sales_data_cleaned.groupby('Category')['Total Sales'].sum()
```

```python
plt.figure(figsize=(8, 8))
colors = plt.cm.Paired(range(len(sales_by_category)))  # Generate a color map for the pie chart
plt.pie(sales_by_category, labels=sales_by_category.index, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Total Sales by Category')
plt.show()
```
